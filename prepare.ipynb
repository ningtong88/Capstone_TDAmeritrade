```python
crsp_db.raw_sql('select a.permco, a.min_date, a.max_date, b.ticker from (select permco, min(date) as min_date, max(date) as max_date from crspm.dsf where permco in %(permco)s group by permco order by 2 asc) a, crsp.stocknames b where a.permco = b.permco', params=permco_list, date_cols=['date'], index_col=['permco'])
```
